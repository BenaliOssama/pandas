```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/household_power_consumption.txt', sep=';')
df.head()
```

```python
df = df.drop(columns=['Time', 'Sub_metering_2', 'Sub_metering_3'])
df.head()
```

```python
df.index = pd.to_datetime(df.index, format='%d/%m/%Y')
df.index
```

```python
df.dtypes
```

```python
def update_types(df): #find the best matching type
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce') # coerce -> nan if can't be converted
    return df

df = update_types(df)
df.dtypes
```

```python
df.describe() # what's the range of values? Are there any obvious outliers? How many non-null values are there?
```

```python
df = df.dropna()
df.describe()
```

```python
df['Sub_metering_1'] = (df['Sub_metering_1'] + 1) * 0.06
df.head()
```
