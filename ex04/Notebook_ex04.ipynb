```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/iris.csv')
df.head()
```

```python
df = df.drop(columns=['flower'])
df.head()
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.dtypes
```

```python
df.isnull().sum()
```

```python


df['sepal_length'] = df['sepal_length'].fillna(df['sepal_length'].mean())
# mdedian is less effected by extraim
df['sepal_width']  = df['sepal_width'].fillna(df['sepal_width'].median())
df['petal_length'] = df['petal_length'].fillna(0)
df['petal_width']  = df['petal_width'].fillna(0)
df.isnull().sum()
```

```python
# Reload fresh data first so we have missing values again
df2 = pd.read_csv('data/iris.csv')
df2 = df2.drop(columns=['flower'])
for col in df2.columns:
    df2[col] = pd.to_numeric(df2[col], errors='coerce')

# Now fill everything with median
df2 = df2.fillna(df2.median())
df2.isnull().sum()
```

```python
#Filling with 0:
#If a flower has a missing petal_length, filling it with 0 tells every future algorithm "this flower has no petal." That's not true — the measurement is just unknown. You've invented a data point that will pull the mean down and distort any model trained on it.
#Filling with the mean:
#Less harmful, but still problematic. Imagine 29 missing petal_length values all get replaced with the same mean value. Now you have 29 identical data points that didn't exist. This artificially reduces the variance of the column — the data looks more "clustered" than it really is, which can mislead machine learning models into false confidence.
```
